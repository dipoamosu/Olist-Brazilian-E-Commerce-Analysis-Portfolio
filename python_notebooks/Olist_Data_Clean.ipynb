```python
#loading every worksheets
df_Location = pd.read_excel("TRANSFORMED DATA.xlsx", sheet_name="customers location")
df_Payment_type = pd.read_excel("TRANSFORMED DATA.xlsx", sheet_name="payment type")
df_high_value_customer = pd.read_excel("TRANSFORMED DATA.xlsx", sheet_name="high value customer")
df_monthly_growth = pd.read_excel("TRANSFORMED DATA.xlsx", sheet_name="monthly growth")
df_number_of_orders = pd.read_excel("TRANSFORMED DATA.xlsx", sheet_name="number of orders")
df_product_category_revenue = pd.read_excel("TRANSFORMED DATA.xlsx", sheet_name="product category revenue")
df_high_revenue_low_sales = pd.read_excel("TRANSFORMED DATA.xlsx", sheet_name="high revenue low sales")
df_sales_over_time = pd.read_excel("TRANSFORMED DATA.xlsx", sheet_name="sales over time")
df_seller_performance = pd.read_excel("TRANSFORMED DATA.xlsx", sheet_name="seller's performance")
```

```python
# Adding ranking and percentages to the products revenue for visualisation and comparison
df_product_category_revenue["Revenue Rank"] = (
    df_product_category_revenue["revenue"]
    .rank(
        ascending=False,
        method="dense"
    )
)

total_revenue = (
    df_product_category_revenue["revenue"]
    .sum()
)


df_product_category_revenue["Revenue Contribution %"] = (
    df_product_category_revenue["revenue"]
    /
    total_revenue
    *100
)
```

```python
df_monthly_growth["Growth %"] = (
    df_monthly_growth["revenue"]
    .pct_change()
    *100
)
df_monthly_growth
```

```python
datasets = {
    "Location": df_Location,
    "Payment_type": df_Payment_type,
    "High_value_customer": df_high_value_customer,
    "Monthly_growth": df_monthly_growth,
    "Number_of_orders": df_number_of_orders,
    "Product_category_revenue": df_product_category_revenue,
    "High_revenue_low_sales": df_high_revenue_low_sales,
    "Sales_over_time": df_sales_over_time,
    "Seller_performance": df_seller_performance
}
```

```python
#checking the column names
for name, df in datasets.items():
    print(name)
    print(df.columns.tolist())
    print()
```

```python
#Removing the spaces in the column names
for name, df in datasets.items():
    df.columns = (
        df.columns
        .str.strip()
        .str.lower()
        .str.replace(" ", "_")
    )
```

```python
#cleaning the text columns
for name, df in datasets.items():

    text_columns = df.select_dtypes(
        include="object"
    ).columns

    for col in text_columns:
        df[col] = (
            df[col]
            .str.strip()
            .str.lower()
        )
```

```python
#fixing the texts with my ftfy library but first i will create a function to clean the text
def clean_text(value):
    if isinstance(value, str):
        value = ftfy.fix_text(value)
        value = value.strip()
    return value
```

```python
#now applying the fix to all my worksheets
for name, df in datasets.items():

    text_columns = df.select_dtypes(
        include="object"
    ).columns

    for col in text_columns:
        df[col] = df[col].apply(clean_text)

    print(f"{name} cleaned successfully")
```

```python
#checking for duplicates
for name, df in datasets.items():

    print(
        name,
        "duplicates:",
        df.duplicated().sum()
    )
```
